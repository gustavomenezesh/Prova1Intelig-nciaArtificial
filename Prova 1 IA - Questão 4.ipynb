## a) Técnicas de pré-preocessamento utilizadas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import cluster, neighbors, svm, metrics, preprocessing

wine_base = pd.read_csv('wine.csv', encoding = "ISO-8859-1")
```

### Verificando se há valores duplicados na base

Pois pode atrapalhar na classificação do método

```python
print('Valores duplicados:', wine_base.duplicated().sum())
```

```python
# Eliminando-os

wine_base.drop_duplicates()
```

### Verificando se há dados faltosos na base
Porque se haver, precisaremos ou completar ou removermos da base para não influenciar negativamente o método que irá ser utilizado

```python
print('Valores faltantes:', wine_base.isnull().sum())
```

Não houve dados faltosos, então não é feito nada

### Transformando a qualidade do vinho(Bom ou Ruim) em Binário(0 ou 1)

```python
le = preprocessing.LabelEncoder()
for column in wine_base.columns:
    if wine_base[column].dtypes == 'object':
        wine_base[column] = le.fit_transform(wine_base[column])
        
print(wine_base)
```

### Montando mapa de correlação entre os atributos

Para poder ser analisado a correlação entre os atributos na base, já que se algum tiver uma correlação forte com algum outro, pode-se retirar da base, pois ambas possuem um comportamento muito parecido.

```python
corr = wine_base.corr()
corr.style.background_gradient(cmap='coolwarm')
```

Não foi necessária a remoção de nenhum atributo, pois não foi obtido nenhuma correlação forte entre eles

## b) Aplicação do método KNN

### Separando base de treinamento e teste

```python
trainSample = wine_base.sample(frac=0.8, random_state = 1) #seleção de 80% da base para treinamento
testSample = pd.concat([wine_base, trainSample]).drop_duplicates(keep=False) #concatena a base original com a base de treinamento e exclui os duplicados, sobrando a amostra de teste

trainTarget = trainSample["quality"] #define o atributo alvo da base para o treinamento
testTarget = testSample["quality"] #define o atributo alvo da base para o teste


del testSample["quality"]
del trainSample["quality"]
```

### Aplicando o método para k=1

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=1, algorithm="auto") 
knn.fit(trainSample, trainTarget)
resultKNN = knn.predict(testSample)

print("Acurácia: %0.2f" % (metrics.accuracy_score(testTarget, resultKNN)))
```

### Para k = 3

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=3, algorithm="auto") 
knn.fit(trainSample, trainTarget)
resultKNN = knn.predict(testSample)

print("Acurácia: %0.2f" % (metrics.accuracy_score(testTarget, resultKNN)))
```

### Para k = 5

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=5, algorithm="auto") 
knn.fit(trainSample, trainTarget)
resultKNN = knn.predict(testSample)

print("Acurácia: %0.2f" % (metrics.accuracy_score(testTarget, resultKNN)))
```

### Para k = 7

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=7, algorithm="auto") 
knn.fit(trainSample, trainTarget)
resultKNN = knn.predict(testSample)
print("Acurácia: %0.2f" % (metrics.accuracy_score(testTarget, resultKNN)))
```

### Para k=9

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=9, algorithm="auto") 
knn.fit(trainSample, trainTarget)
resultKNN = knn.predict(testSample)

print("Acurácia: %0.2f" % (metrics.accuracy_score(testTarget, resultKNN)))
```

## c) Analisando o comportamento do método com métricas de distâncias diferentes
